```python
# Install yfinance library to fetch stock data
!pip install yfinance

# Import the yfinance library to get financial data
import yfinance as yf

# Import matplotlib to create charts
import matplotlib.pyplot as plt

# Get Apple stock data from the last year
stock = yf.Ticker("AAPL")
# Download 1 year of historical data
data = stock.history(period="1y")

# Show the first 5 rows of data
data.head()

# Plot the closing price
plt.figure(figsize=(10, 5))
plt.plot(data['Close'], label="Closing Price")

# Add chart title and labels
plt.title("Apple (AAPL) Stock Price - Last 1 Year")
plt.xlabel("Date")
plt.ylabel("Price (USD)")

# Add a legend and grid for better readability
plt.legend()
plt.grid(True)

# Show the final plot
plt.show()


# Calculate the 50-day moving average of closing prices
data['MA50'] = data['Close'].rolling(window=50).mean()

# Plot the closing price and moving average together
plt.figure(figsize=(10, 5))
plt.plot(data['Close'], label="Closing Price")
plt.plot(data['MA50'], label="50-Day Moving Average", color='orange')

# Add titles, labels, legend
plt.title("Apple Stock with 50-Day Moving Average")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()





```
